```python
import pandas as pd
import statistics
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("complete_results_greedy_200km_join2_airport3_maxsteps3700.txt")
```

```python
df1.plot(y='Total Fuel Used', style='o')
```

```python
ax = df1.plot( y='Total Fuel Used', style='o')
df1.plot( y="steps", style='o', ax=ax)
df1.plot( y="Av Savings Alliance", style='o', ax=ax)


#df.plot( y="C", style='o', ax=ax, color="C3")
```

```python
df2 = df1[['Total Fuel Used', 'steps']] 
```

```python
TFU = []
mean =[]
std = []
coffvarTFU = []

for index, row in df2.iterrows():
    TFU.append(df2['Total Fuel Used'][index])
    if index > 2:
        a = sum(TFU)/len(TFU)
        mean.append(a)
        b = statistics.stdev(TFU)
        std.append(b)
        coffvarTFU.append(b/a)
print(coffvarTFU)    


steps = []
mean =[]
std = []
coffvarsteps = []
    
for index, row in df2.iterrows():
    steps.append(df2['steps'][index])
    if index > 2:
        a = sum(steps)/len(steps)
        mean.append(a)
        b = statistics.stdev(steps)
        std.append(b)
        coffvarsteps.append(b/a)
print(coffvarsteps)    

    
#steps = []    
#steps.append(df2['steps'][index])    
#print(steps)    
```

```python
plt.plot(range(len(coffvarTFU)), coffvarTFU)
plt.ylabel('Coefficient of variance Total Fuel Used')
plt.xlabel('Iteration')
plt.show()
```

```python
plt.plot(range(len(coffvarsteps)), coffvarsteps)
plt.ylabel('Coefficient of variance steps')
plt.xlabel('Iteration')
plt.show()
```
